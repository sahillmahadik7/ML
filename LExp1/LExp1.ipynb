```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("train.csv")

# View first 5 rows
df.head()
```

```python
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
target = "Survived"

X = df[features]
y = df[target]
```

```python
# Fill missing Age with median
X["Age"] = X["Age"].fillna(X["Age"].median())

# Fill missing Embarked with most frequent value
X["Embarked"] = X["Embarked"].fillna(X["Embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
X.isnull().sum()
```

```python
# Convert Sex column
le_sex = LabelEncoder()
X["Sex"] = le_sex.fit_transform(X["Sex"])

# Convert Embarked column
le_embarked = LabelEncoder()
X["Embarked"] = le_embarked.fit_transform(X["Embarked"])
```

```python
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X, y)
```

```python
test_df = pd.read_csv("test.csv")

# Check first 5 rows
test_df.head()
```

```python
X_test = test_df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
```

```python
X_test.isnull().sum()
```

```python
X_test["Age"] = X_test["Age"].fillna(X_test["Age"].median())
X_test["Fare"] = X_test["Fare"].fillna(X_test["Fare"].median())
X_test["Embarked"] = X_test["Embarked"].fillna(X_test["Embarked"].mode()[0])
```

```python
# Use the SAME encoders fitted on train data
X_test["Sex"] = le_sex.transform(X_test["Sex"])
X_test["Embarked"] = le_embarked.transform(X_test["Embarked"])
```

```python
test_predictions = model.predict(X_test)
```

```python
submission = pd.DataFrame({
    "PassengerId": test_df["PassengerId"],
    "Survived": test_predictions
})

submission.head()
```
